```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\hugoe\Desktop\LABS\lab-eda-univariate\archive\amz_uk_price_prediction_dataset.csv")
df.head()
```

```python
category_counts = df["category"].value_counts()
category_counts
```

```python
top_categories = category_counts.head(5)
top_categories
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.title("Distribucin de Productos en las Principales Categoras")
plt.xlabel("Categora de Producto")
plt.ylabel("Frecuencia")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_categories, labels=top_categories.index, autopct="%1.1f%%")
plt.title("Proporcin de Productos en las Principales Categoras")
plt.show()
```

```python
mean_price = df["price"].mean().round()
mean_price 
```

```python
median_price = df["price"].median()
median_price 
```

```python
mode_price = df["price"].mode()[0]
mode_price 
```

```python
price_variance = round(df["price"].var(),2)
price_variance
```

```python
price_std_dev = round( df["price"].std(),2)
price_std_dev
```

```python
price_range = df["price"].max() - df["price"].min()
price_range
```

```python
iqr_price = df["price"].quantile(0.75) - df["price"].quantile(0.25)
iqr_price
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["price"], bins=30, kde=True)
plt.title("Distribucin de Precios de Productos")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["price"])
plt.title("Box Plot de Precios de Productos")
plt.xlabel("Precio")
plt.show()
```
